```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

datasets = [
    "predictions_smard_real_lstm_seq2seq_additive_corrected_hour",
    "predictions_smard_real_mse_lstm_naive_hour",
    "predictions_smard_real_mse_lstm_seq2seq_additive_regu_31_07_hour",
    "predictions_smard_real_mse_lstm_stacked_hour",
]

for dataset in datasets:
    data = pd.read_csv(f"{dataset}.csv")
    data["imbalance_pred"] = data["pred_total_production_mwh"] - data["pred_total_load_mwh"]
    data["imbalance_real"] = data["total_production_mwh"] - data["total_load_mwh"]
    data["imbalance_smard"] = data["smard_total_production_mwh"] - data["smard_total_load_mwh"]
    data.to_csv(f"imbalance_{dataset}.csv", index=False)
```

```python
data["imbalance_real"].plot(ylim=(-11000, 11000))
```

```python
data["imbalance_pred"].plot()
```
